```python
import matplotlib.pyplot as plt

# Data
agents = ['Deep Q Learning', 'Double Deep Q Learning', 'Q Learning', 'SARSA', 'Random Agent']
reward_means = [2320.93, 2314.11, 1074.54, 1861.225, 1632.196]
reward_stds = [381.50, 386.19, 112.92, 358.469, 356.37]

# Plot
plt.figure(figsize=(10, 6))
plt.bar(agents, reward_means, yerr=reward_stds, capsize=5)
plt.xlabel('Agent')
plt.ylabel('Reward Mean')
plt.title('Reward Mean and Standard Deviation for Each Agent')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
